```python
"""
How do I solve the drunkard's walk or random walk simulation using Python language?


The "drunkard's walk" or "random walk" isn't exactly something to be solved,
but rather is a random process that we can simulate. The process has one step taken
in a random direction at each discrete unit of time.

In a simple, famous example, the process can be metaphorically described by a
drunkard trying to get home. Due to his (or her!) inebriated status, he walks
randomly as opposed to in the direction of his home. We might start in the simple
case of one dimension, which we can think of as the drunk being on a street, and
at each discrete timestep, either takes a step forward on the street or a step backward.

---

Without numpy:
"""
```

```python
import random
import pandas as pd
import matplotlib.pyplot as plt
 
position = 0
walk = [position]
for step in range(100):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
 
# plt.figure(figsize=(10,5))
plt.plot(walk, color= "crimson", linestyle="dashdot")
```

```python
"""
With numpy:
"""
```

```python
import numpy as np
def display():
    draws= np.random.randint(0, 2, size=100)
    steps = np.where(draws > 0, 1, -1)
    walk = steps.cumsum()

    # plt.figure(figsize=(10,5))
    plt.plot(walk, color= "blue", linestyle="dashdot" )

display()
```
